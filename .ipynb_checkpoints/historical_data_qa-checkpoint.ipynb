```julia
using CSV
using DataFrames
using DataFramesMeta
using Dates
using Distributions
using HypothesisTests
using Plots
using Statistics
using StatsPlots
```

```julia
folder = joinpath(@__DIR__(), "data")
df = CSV.read(joinpath(folder, "nanaimo_ridership_fall_2023.csv"), DataFrames.DataFrame);
```

```julia
first(df, 5)
```

```julia
df = df[df.route .!= 70, :];
```

```julia
holidays = [
    Date(2023, 9, 4), # labour day
    Date(2023, 9, 30), # truth and reconciliation
    Date(2023, 10, 9), # thanksgiving
    Date(2023, 11, 11), # remembrance day
]

df.is_holiday = [date in holidays for date in df.date];
```

```julia
df.is_weekday = Dates.dayofweek.(df.date) .< 5;
```

```julia
df.primary_delay_seconds = df.end_delay_seconds - df.start_delay_seconds;
```

```julia
df.primary_delay_pct = df.primary_delay_seconds ./ df.planned_duration_seconds;
```

```julia
df.planned_start_hour = Dates.hour.(df.planned_start_time);
```

```julia
x = df.primary_delay_pct[.!df.is_holiday .& df.is_weekday];
```

```julia
fit = Distributions.fit(Distributions.Normal, x)
```

```julia
histogram(
    x,
    bins=range(-1, 1, length=51),
    normalize=true,
    xlabel="% of trip length",
    ylabel="count",
    title="Primary delays for weekday trips",
    legend=nothing
)
plot!(
    fit
)
```

```julia
# Kolmogorov-Smirnov test
HypothesisTests.ExactOneSampleKSTest(x, fit)
```

```julia
# Anderson-Darling test
HypothesisTests.OneSampleADTest(x, fit)
```

```julia
weekday_df = df[.!df.is_holiday .& df.is_weekday, :];
```

```julia
route_time_direction_df = DataFrames.groupby(weekday_df, [:route, :planned_start_hour, :direction]);
time_direction_df = DataFrames.groupby(weekday_df, [:planned_start_hour, :direction]);
```

```julia
combined_df = combine(
    route_time_direction_df,
    :total_boardings => mean => :ridership,
    :primary_delay_seconds => mean => :μ,
    :primary_delay_seconds => std => :σ
)
```

```julia
@df combined_df[combined_df.route .== 40, :] plot(
    :planned_start_hour,
    :μ,
    ribbon=:σ,
    fillalpha=0.2,
    group=:direction
)
```
