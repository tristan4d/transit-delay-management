```julia
# Loads a trips dataframe and shapes dataframe
# Can use Nanaimo or Cranbrook, but Nanaimo is more interesting I believe as it has more varied terminus locations

trips, shapes = loadGTFS("..\\data\\Nanaimo-GTFS");
# trips, shapes = loadGTFS("..\\data\\Cranbrook-GTFS");
```

```julia
# Create a subset of the trips dataframe with 30 trips that start between 0800-1200
subset = subsetGTFS(trips; n = 30, start_time = 8, stop_time = 12, randomSeed = 1)

# Generate normal distribution means and stds for each trip in the subset, can change form, mean, and std
l = primaryDelays(subset; form = 0, meanMulti = 1.0, stdMulti = 1.0) # l = [μ σ]

# Create the problem instance
instance = VSPInstance(subset; l = l)

# Create the JuMP model, includes sampling the distribution for each trip numScenario times
# Split controls how many scenarios are considered in the optimization
del_model = VSPModel(instance; numScenarios = 100, split = 1.0, randomSeed = 1, timeLimit = 60)

# A min-cost flow model for comparison
mcf_model = MCFModel(instance)

# Get solutions
del_solution = solve!(del_model; silent = false)
mcf_solution = solve!(mcf_model);
```

```julia
# Plot min-cost solution using the training delays to visualize propagated delay
mcf_plot_time = plotVSP_time(mcf_solution; delays = del_model.L_train)
```

```julia
# Plot the min-cost-delay solution using the training delays to visualize propagate delay
del_plot_time = plotVSP_time(del_solution; delays = del_model.L_train)
# del_plot_time = plotVSP_time(del_solution; delays = del_model.L_test)
```

```julia
# Get metrics for each model

del_stats_train = getSolutionStats(del_solution, shapes, del_model.L_train)
# del_stats_test = getSolutionStats(del_solution, shapes, del_model.L_test) # commented because L_test is empty with split = 1.0
mcf_stats = getSolutionStats(mcf_solution, shapes, del_model.L_train);
```
