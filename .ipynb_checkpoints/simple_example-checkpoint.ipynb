```julia
using Distributions
using Plots
using QuadGK
```

```julia
dist = Distributions.Normal(0, 5)
veh_cost = 600
op_cost = 160
delay_cost = 37
ridership = 10:10:100
buffers = 2:2:20
```

```julia
rta = zeros(Bool, 10, 10)

for i in 1:10
    for j in 1:10
        result[i, j] = quantile(dist, 1 - op_cost / delay_cost / ridership[i]) > buffers[j]
    end
end
```

```julia
heatmap(
    buffers,
    ridership,
    rta,
    color=:bluesreds,
    colorbar=nothing
)
```

```julia
delay_aware = zeros(Bool, 10, 10)
f(x) = x * pdf(dist, x)

for i in 1:10
    integral, error = quadgk(f, buffers[i], Inf)
    d = integral * (1 - cdf(dist, buffers[i]))
    @show d
    for j in 1:10
        delay_aware[j, i] = delay_cost * ridership[j] * d > veh_cost
    end
end
```

```julia
delay_aware
```

```julia
heatmap(
    buffers,
    ridership,
    delay_aware,
    color=:bluesreds,
    colorbar=nothing
)
```
